```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/Satish-kumar26/Python-ca/main/big12basketball.csv")
df.head()
```

**Breakdown of Rank 1-12**

```python
df["defensive_rebounds"]=df["total_rebounds"]-df["offensive_rebounds"]
for y in ["net_rating", "points","field_goal_percentage","3_pointer_percentage","free_throw_percentage","total_rebounds","offensive_rebounds","defensive_rebounds","assists","steals","blocks","turnovers","personal_fouls"]:
  fig=px.box(df, x="rank",y=y,notched=True,color_discrete_sequence=px.colors.qualitative.Bold, hover_data =["school"], color="rank", title = f" {y} Per Game - Rank 1-12")

  fig.update_layout(template = "plotly_dark", font = dict(family = "PT Sans", size=16, color="#FF9A00"))
  fig.update_traces(fillcolor="#000000")
  fig.update_traces(marker=dict(size=8,line=dict(width=1, color ="#FFFFFF")))
  fig.show()
```

**Points Per Game by Year and Big 12 Ranking**

```python
fig = px.scatter(df, x ="rank", y = "year", hover_data = ["school"],
                color = "points", color_continuous_scale = "Jet",
                title= "Points Per Game by Year and Big 12 Ranking")
fig.update_traces(marker=dict(size =11, symbol = "circle"))
fig.update_layout(template = "plotly_dark", font =dict(family = "PT Sans", size=20))
fig.show()
```

**Big 12 Conference Ranking by Year**

```python
for count, school in enumerate (sorted(['Baylor', 'Colorado', 'Iowa State', 'Kansas', 'Kansas State', 'Missouri', 'Nebraska', 'Oklahoma', 'Oklahoma State', 'TCU', 'Texas', 'Texas A&M', 'Texas Tech', 'West Virginia'])):
  df0=df[df["school"]==school]
  fig=px.line(df0,x="year",y="rank",markers="true",text="rank",
              hover_name="rank",hover_data=["school"],
              title=f"{school} Basketball Big 12 Conference Ranking by Year")
  fig.update_traces( textposition="top center")
  fig.update_traces(marker=dict(size=10, line=dict(width=1,color="#FFFFFF")))
  fig.update_layout(template="plotly_dark",font=dict(family="PT Sans", size=15,color="#FF9A00"))
  fig.show()
```

**Home Wins and Home Losses by Year**

```python
for count, school in enumerate(sorted(['Baylor', 'Kansas', 'Texas Tech', 'Texas', 'Oklahoma State', 'TCU',
 'Iowa State', 'Oklahoma', 'Kansas State', 'West Virginia', 'Missouri','Texas A&M', 'Colorado', 'Nebraska'])):
    
    df0 = df[df["school"] == school]
    
    fig = px.bar(df0, x = "year",
                 y = ["home_wins", "home_losses"],
                 color_discrete_sequence = ["#009700", "#970000"],
                 hover_name = "rank", hover_data = ["school"],
                 title = f"{school} Basketball Home Wins and Home Losses by Year")

    fig.update_layout(template = "plotly_dark", font = dict(family = "PT Sans", size = 16, color = "#FF9A00"))

    fig.show()
```
